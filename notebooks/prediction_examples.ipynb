# Prediction Examples

```python
import json
import random
from PIL import Image
```

## Open predictions and reference file

```python
with open("../output/pred_captions_ViT_MLP_both.json", "r") as f:
    preds = json.load(f)

with open("../deepfashion-mm/captions.json", "r") as f:
    ref_captions = json.load(f)
    
```

## Sampling of random result

```python
pred = random.sample(preds, 1)[0]

f = f'../deepfashion-mm/images/{pred["image_id"]}'

segm = Image.open(f)

display(f"FILENAME: {pred['image_id']}")
display(f"PREDICTED CAPTION: {pred['caption']}")
display(f"REFERENCE CAPTION: {ref_captions[pred['image_id']]}")
display(segm)
```
